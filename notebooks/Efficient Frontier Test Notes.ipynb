```python
from qfntools import EfficientFrontier
import pandas as pd
import numpy as np

# Preferably weekly or monthly
data = pd.read_excel('./data/biggestETFData.xlsx',index_col = 0)
data = data.resample('M').last()
data_return= data.pct_change().dropna(how = 'all')
```

```python
ef = EfficientFrontier(risk_measure='entropy', alpha=5, entropy_bins=20)
ef.fit(data_return, wbnd=(0,1), mu_range=np.arange(0.0055,0.01,0.0002))
```

```python
df = ef.weights()
df.columns = ['Stock{}'.format(i) for i in range(1, df.shape[1]+1)]
df
```

```python
ef.plot(port_only=False)
```
